# Multivariate analysis:

```python
#Data Cleaning --> Duplicate values and we are going to treat null values.
#Feature engineering --> creation of feature based on existing ones.
#Data anaysis --> performing analysis on data based on diffrent factors.
```

```python
#Duplicate values: 

df.duplicated().sum()
```

```python
#What if it has duplicate set of values?

df=df.drop_duplicated()
```

```python
df
```

```python
#null values:
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["State"]=df["State"].fillna(df["State"].mode()[0])
```

```python
df["State"].mode()[0]
```

```python
df["Deaths"]=df["Deaths"].fillna(df["Deaths"].median)
```

# For Loop:

```python
list_1 = [10,20,30,40]

for i in list_1:
    print(i)
```

```python
set_1={10,20,30,40,50,10}
for i in set_1:
    print(i)
```

```python
dict_1={"name":"Dhaval","age":24,"role":"LLMops"}
for i in dict_1:
    print(i)
```

```python
for j in dict_1.values():
    print(j)
```

```python
for k in dict_1.items():
    print(k)
```

```python
#to find put even number from first natural number

for i in range(1,11):
    print(i)
```

```python
for i in range(2,21,2):
    print(i)
```

```python
list_1=[10,20,30,40]
for i in enumerate(list_1):
    print(i)
```

```python
list_2=[11,22,33,44]
for i,j in zip(list_1,list_2):
    print(i+j)
```

# While loop:

```python
x=1

while x<= 10:
    print(x)
    x+=1
```

```python
#print all the numbers entered until user enter the number 0
num=int(input("Enter the number:"))

while num!=0:
    print(num)
    num=int(input("Enter the next number:"))
    
print("End")
```

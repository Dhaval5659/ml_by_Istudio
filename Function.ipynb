```python
#sum of two numbers using functions:

def sum_1(num_1,num_2):
    return num_1+num_2
```

```python
sum_1(10,20)
```

```python
# when to print and when to use return:

def sub_1(a,b):
    return a-b
```

```python
def master (a1,b1):
    ans_1=sum_1(a1,b1)
    ans_2=sub_1(a1,b1)
    print(ans_1,ans_2)
```

```python
master(45,85)
```

```python
#variable length argument.

def sum_n(*param_1):
    sum_1=0
    for i in param_1:
        sum_1+=i
    print(sum_1)
```

```python
sum_n(10,20,30)
```

```python
sum_n(10,20,30,50,60)
```

```python
#cube of a number:

num_cube=lambda a:a**3
num_cube(10)
```

## Recursive functions: 

```python
Syntax: 
 
def name_of_functions(parameter_1,.....):
    .
    .
    .
    .
    name_of_functions(arg,......)
    
call:
name_of_function(arg,...)
```

```python
def factorial(x):
    if x==1:
        return 1
    else:
        return x * factorial(x-1)
```

```python
factorial(5)
```

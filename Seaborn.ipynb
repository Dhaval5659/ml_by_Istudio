```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/covid_19_data.csv")
```

```python
df
```

```python
#univariate categorical
sns.catplot(x="Deaths",data=df,kind="bar")
```

```python
help(sns.catplot)
```

```python
#univariable numerical feature:
plt.figure(figsize=(10,5))

sns.distplot(df["Deaths"],hist=False)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Deaths"])
```

```python
df.dtypes
```

```python
#bivariate numerical:

sns.jointplot(x="Confirmed",y="Deaths",data=df,kind="scatter",hue="Recovered")
```

```python
sns.pairplot(data=df)
```

```python
#bivariate categorical

sns.boxplot(x="State",y="Deaths",data=df)
```

# Stacking and Spliting of a matrix  

```python
mat_1=np.arange(12).reshape(3,4)
mat_2=np.arange(12,24).reshape(3,4)
```

```python
mat_1
```

```python
mat_2
```

```python
v=np.vstack((mat_1,mat_2))
v
```

```python
h=np.hstack((mat_1,mat_2))
h
```

```python
np.hsplit(v,2)
```

```python
np.vsplit(h,3)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[6,5,7,8,9]

plt.plot(x,y,marker="o",color="b")
```

```python
help(plt.plot)
```

```python
#Bar plot

x_1=[5,2,9,47]
y_1=[10,20,35,45]

plt.bar(x_1,y_1,color=["red","green","pink","brown"])

plt.xlabel("Sales")
plt.ylabel("Profit")

plt.title("Sales portfolio")
```

```python
help(plt.bar)
```

```python
#histogram

axis_1=[10,20,30,40]
plt.hist(y)
```

```python
#scatter plot

x=[10,20,30,41]
y=[10,5,6,7]

plt.scatter(x,y,color="r")
```

```python
#stacked plot
var_1=["A","B","C","D"]
y1=[10,20,30,40]
y2=[2,5,3,5]

plt.bar(var_1,y1,color="r")
plt.bar(var_1,y2,color="b")
```

```python
#pie chart
label_1=["man","woman","other","undefined"]
y_n=np.array([25,35,65,40])
plt.pie(y_n,labels=label_1,explode=[0.5,0,0.2,0])
plt.legend()
```
